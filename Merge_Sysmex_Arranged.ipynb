```python
from google.colab import drive

# Specify the folder path containing the Excel files
folder_path = '/content/drive/MyDrive/4-24-2024_4 outputs'

# Import pandas
import pandas as pd

# Read the three excel sheets into Pandas DataFrames
df1 = pd.read_csv(f'{folder_path}/XN_00-22 (Build 182)_SAMPLE.csv')
df2 = pd.read_csv(f'{folder_path}/XN_00-22 (Build 182)_SAMPLE_2.csv')
df3 = pd.read_csv(f'{folder_path}/XN_00-22 (Build 182)_SAMPLE_3.csv')
```

```python
# drop all columns with only NaN contents (all columns with empty values)

df1.dropna(axis=1, how='all', inplace=True)

df2.dropna(axis=1, how='all', inplace=True)

df3.dropna(axis=1, how='all', inplace=True)
```

```python
num_columns = len(df1.columns)
# Insert the column from df2 into df1 as the last column
df1.insert(loc=num_columns, column="IPF#(10^3/uL)", value=df2["IPF#(10^3/uL)"])
```

```python
# Move entire contents of the 9 columns from df3 and add to end of df1

last_9_column_titles = df3.columns[-9:]

last_9_column_titles

df1[last_9_column_titles] = df3[last_9_column_titles]
```

```python
 # drop columns that are not needed if they exist in df1

 def drop_columns(df1, columns_to_drop):

  for column in columns_to_drop:
    if column in df1.columns:
      df1.drop(column, axis=1, inplace=True)
    return df1

columns_to_drop = ['Nickname', 'Time', 'Analyzer ID', 'Rack', 'Sample Inf.', 'Order Type', 'Reception Date', 'Measurement Mode', 'Discrete', 'Analysis Info.', 'Judgment', 'Positive(Count)', 'PLT Suspect', 'WBC Info.', 'PLT Info.', 'Validate', 'Validator', 'Action Message (Review)', 'Action Message (Retest)', 'Output', 'Sequence No.', 'IP SUS(PLT)PLT Clumps?', 'Q-Flag(Blasts/Abn Lympho?)', 'Q-Flag(Blasts?)', 'Q-Flag(Abn Lympho?)', 'Q-Flag(Left Shift?)', 'Q-Flag(Atypical Lympho?)', 'Q-Flag(RBC Agglutination?)', 'Q-Flag(Turbidity/HGB Interf?)', 'Q-Flag(Iron Deficiency?)', 'Q-Flag(HGB Defect?)', 'Q-Flag(Fragments?)', 'Q-Flag(PLT Clumps?)', 'WBC/M', 'RBC/M', 'HGB/M', 'HCT/M', 'PLT/M', 'MPV/M', 'NEUT#/M', 'LYMPH#/M', 'MONO%/M', 'BASO%/M', 'IPF/M', 'CELLPACK DCL', 'SULFOLYSER', 'Lysercell WNR', 'Fluorocell WNR', 'Lysercell WDF', 'Fluorocell WDF', 'CELLPACK DFL', 'Fluorocell RET', 'Fluorocell PLT']

df1 = drop_columns(df1, columns_to_drop)
```

```python
df1.columns
```

```python
# Rename the column 'Date' to 'Testing Date'
df1.rename(columns={'Date': 'Testing Date'}, inplace=True)
```

```python
# Remove indent from the "Sample No." column
df1['Sample No.'] = df1['Sample No.'].str.replace('      =', '')
```

```python
# Change order of columns
new_order = ['Testing Date','Time','Sample No.','WBC(10^3/uL)', 'RBC(10^6/uL)', 'HGB(g/dL)', 'HCT(%)', 'MCV(fL)', 'MCH(pg)', 'MCHC(g/dL)', 'RDW-SD(fL)', 'RDW-CV(%)', 'PLT(10^3/uL)', 'MPV(fL)', 'NEUT#(10^3/uL)', 'NEUT%(%)', 'LYMPH#(10^3/uL)', 'LYMPH%(%)', 'MONO#(10^3/uL)', 'MONO%(%)', 'EO#(10^3/uL)', 'EO%(%)', 'BASO#(10^3/uL)', 'BASO%(%)', 'IG#(10^3/uL)', 'IG%(%)', 'NRBC#(10^3/uL)', 'NRBC%(%)', 'RET%(%)', 'RET#(10^6/uL)', 'IRF(%)', 'RET-He(pg)', 'IPF(%)', 'IPF#(10^3/uL)']
df1 = df1.reindex(columns=new_order)
```

```python
# Save the merged DataFrame df1 to a new Excel file

from datetime import date

today = date.today().strftime('%m-%d-%Y')

df1.to_excel(f'{folder_path}/merged_file_{today}.xlsx', index=False)
```

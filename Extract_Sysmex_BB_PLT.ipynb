```python
from google.colab import drive

# Specify the folder path containing the Excel files
folder_path = '/content/drive/MyDrive/4-24-2024_BB+PLT_10 outputs'

# Import pandas
import pandas as pd

# Read the three excel sheets into Pandas DataFrames
df1 = pd.read_csv(f'{folder_path}/XN_00-22 (Build 182)_SAMPLE_2.csv')
df2 = pd.DataFrame()
```

```python
# drop all columns with only NaN contents (all columns with empty values)

df1.dropna(axis=1, how='all', inplace=True)
```

```python
# Create a list of column names to insert from df1 to df2
columns_to_insert = ["Date","Time","Sample No.","PLT(10^3/uL)"]

# Insert columns in df2
for column in columns_to_insert:
    df2.insert(loc=len(df2.columns), column=column, value=df1[column])
```

```python
# view final DataFrame before saving

df2
```

```python
# Save the merged DataFrame df1 to a new Excel file

from datetime import date

today = date.today().strftime('%m-%d-%Y')

df2.to_excel(f'{folder_path}/extracted_file_BB+PLT_{today}.xlsx', index=False)
```

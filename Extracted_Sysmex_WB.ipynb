```python
from google.colab import drive

# Specify the folder path containing the Excel files
folder_path = '/content/drive/MyDrive/4-24-2024_WB_10_only PLT'

# Import pandas
import pandas as pd

# Read the three excel sheets into Pandas DataFrames
df1 = pd.read_csv(f'{folder_path}/XN_00-22 (Build 182)_SAMPLE.csv')
df2 = pd.read_csv(f'{folder_path}/XN_00-22 (Build 182)_SAMPLE_2.csv')
df3 = pd.DataFrame()
```

```python
# drop all columns with only NaN contents (all columns with empty values)

#df1.dropna(axis=1, how='all', inplace=True)
#df2.dropna(axis=1, how='all', inplace=True)
```

```python
num_columns = len(df1.columns)
# Insert the column from df2 into df1 as the last column
df1.insert(loc=num_columns, column="IPF#(10^3/uL)", value=df2["IPF#(10^3/uL)"])
```

```python
# Create a list of column names to insert from df1 to df2
columns_to_insert = ['Date','Time','Sample No.','WBC(10^3/uL)','RBC(10^6/uL)','HGB(g/dL)','HCT(%)','MCV(fL)','MCH(pg)','MCHC(g/dL)','PLT(10^3/uL)','RDW-SD(fL)','RDW-CV(%)','MPV(fL)','NRBC#(10^3/uL)','NRBC%(%)','IPF(%)','IPF#(10^3/uL)']

# Insert columns in df2
#for column in columns_to_insert:
   # df3.insert(loc=len(df1.columns), column=column, value=df1[column])

for i, column in enumerate(columns_to_insert):
    df3.insert(loc=i, column=column, value=df1[column])
```

```python
# Save the merged DataFrame df1 to a new Excel file

from datetime import date

today = date.today().strftime('%m-%d-%Y')

df3.to_excel(f'{folder_path}/extracted_file_WB_{today}.xlsx', index=False)
```

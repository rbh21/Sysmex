* Columns with non-empty values from sample 2 and 3 are added to sample 1 (df1)
* Includes only relevant data from output files Sample 2 and Sample 3 which includes IPF#(10^3/uL), IPF#/M from "Sample 2" and last 9 Cell pack columns from "Sample 3"

```python
from google.colab import drive

# Specify the folder path containing the Excel files
folder_path = '/content/drive/MyDrive/9 outputs'

import pandas as pd

# Read the three excel sheets into Pandas DataFrames
df1 = pd.read_csv(f'{folder_path}/XN_00-22 (Build 182)_SAMPLE.csv')
df2 = pd.read_csv(f'{folder_path}/XN_00-22 (Build 182)_SAMPLE_2.csv')
df3 = pd.read_csv(f'{folder_path}/XN_00-22 (Build 182)_SAMPLE_3.csv')
```

```python
df1.dropna(axis=1, how='all', inplace=True)
```

```python
len(df1.columns)
```

```python
df2.dropna(axis=1, how='all', inplace=True)

# Save the new DataFrame to a new Excel sheet
# # df2.to_excel(f'{folder_path}/new_file.xlsx', index=False)
```

```python
num_columns = len(df1.columns)

# Insert the column from df2 into df1 as the last column
df1.insert(loc=num_columns, column="IPF#(10^3/uL)", value=df2["IPF#(10^3/uL)"])

num_columns = len(df1.columns)
# Insert second  column from df2 into df1 as the last column
df1.insert(loc=num_columns, column="IPF#/M", value=df2["IPF#/M"])
```

```python
df3.dropna(axis=1, how='all', inplace=True)
```

```python
# get names of columns of interest from df3

last_9_column_titles = df3.columns[-9:]

last_9_column_titles
```

```python
# Move entire columns with these names from df3 to end of df1
df1[last_9_column_titles] = df3[last_9_column_titles]
```

```python
# check if all columns are correctly added

print(df1.columns)

# number of df1 columns should be equal to 76+2+9=87
len(df1.columns)
```
